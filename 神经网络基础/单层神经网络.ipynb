### 手动实现一个线性回归的神经网络


输入为x1,x2<br>
x0表示b的参数全为1<br>
目的为拟合x和z的曲线<br>
这里的z相当于线性曲线的y，但在神经网络中，一般不规定为y<br>
<br>



```python
import torch
```

```python

# 线性回归任务：拟合曲线
# 输出 = 输入矩阵 × 权重矩阵
# 下面的代码需要注意
# 以后敲代码养成规范养成规范都规定为float32



# 所有前面的1是为了×偏置b
input = torch.tensor([[1,0,0],
                      [1,1,0],
                      [1,0,1],
                      [1,1,1]],dtype=torch.float32)
w = torch.tensor([1,-0.2,0.3],dtype=torch.float32)
print(w.shape)
z = torch.tensor([-0.2,-0.05,-0.05,0.1],dtype = torch.float32)
print(input.type())
print(w.type())
def linearR (x,w):
    output = torch.mv(x,w)
    return output

result = linearR(input,w)
print(result)
print(result.shape)

```

## 使用torch库实现线性回归

```python
# 对上式使用torch.nn.Linear()进行验证
# 调用torch.nn.Linear()w和b参数不用人为设置，pytorch会自动初始化w和b,可以查看随机设置的w和b
#  torch.nn.Linear(2,1)表示输入为两个神经元，输出为1个，输入的两个神经元为x1,x2

input = torch.tensor([[1,0],[1,1],[1,0],[1,1]],dtype=torch.float32)
output = torch.nn.Linear(2,1)
result = output(input)
print(output.weight)
print(output.bias)
print(result)
```

## 二分类问题

```python
# 将线性回归转化为二分类问题
torch.manual_seed(234) # 设置随机种子，保证每次运行结果一致
input = torch.tensor([[0,0],[0,1],[1,1]],dtype=torch.float32)
z = torch.nn.Linear(2,1)
output = torch.sigmoid(z(input))
print(output) # 输出为0-1之间的值
# 这里的输出值可以看作是分类的概率
# 转化为二分类的0,1
output1 =  [int(x) for x in output>=0.5]
print(output1) # 输出为0,1
```
